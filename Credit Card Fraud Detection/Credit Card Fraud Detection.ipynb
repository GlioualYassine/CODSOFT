### Evaluating the Model 

### Classification Metrics:
1. Accuracy:
      - Signification: Proportion of correctly classified instances.
      - Higher Value Signifies: Better overall performance.
2. Precision:
      - Signification: Proportion of true positive predictions among all positive predictions.
      - Higher Value Signifies: Fewer false positives.
3. Recall (Sensitivity or True Positive Rate):
      - Signification: Proportion of true positive predictions among all actual positives.
      - Higher Value Signifies: Fewer false negatives.
4. F1 Score:
      - Signification: Harmonic mean of precision and recall.
      - Higher Value Signifies: A balance between precision and recall.

## ⚠️ Caution: The Danger of Using Accuracy Alone for Imbalanced Datasets

Remember that relying solely on accuracy as a metric for imbalanced datasets can be misleading. In scenarios where one class significantly outnumbers the others, accuracy might not accurately reflect the model's performance. This is because a model can achieve high accuracy by simply predicting the majority class.

It's essential to consider alternative metrics such as precision, recall, F1-score, or area under the ROC curve (AUC-ROC) to gain a more comprehensive understanding of your model's performance, especially in imbalanced classification tasks.


## Handling Imbalanced Dataset

- we can use two way to handle imbalanced dataset : 
    - Undersampling : we are randomly delete rows from majority class to match them with the minority class
    - Oversampling :  

### Undersampling

```python
normal = df[df['Class']== 0 ]
fraud = df[df['Class']== 1 ]
```

```python
normal.shape
```

```python
fraud.shape
```

* normal.sample(n=473) : will select 473 randomly from normal transactions

```python
normal_sample = normal.sample(n=473)
normal_sample.shape
```

* ignore_index=True : will reset the index to start from 0 (default is false which means the random index of the old df will remain)

```python
new_df = pd.concat([normal_sample,fraud],ignore_index=True)
new_df['Class'].value_counts()
```

```python
new_df.head()
```

```python
# splitting data
x = new_df.drop('Class',axis=1)
y = new_df['Class']
```

```python
# splitting the data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=43)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
```

## Oversampling

* the advantage of undersampling is that we lose a lot of valuable data so that's why we are going to use oversampling
* we will use SMOTE
* SMOTE (Synthetic Minority Over-sampling Technique) : try to balance the distribution by randomly increasing minority class , it do that using a linear interpolation 
* Creating Synthetic Samples: For each minority class instance, SMOTE generates synthetic samples by interpolating between that instance and its nearest neighbors. It creates new instances along the line segments connecting the original minority class instances.
*  the avantage of SMOTE is you are not duplication element but rather you are creating synthetic data point that are slightly different from the original data points
